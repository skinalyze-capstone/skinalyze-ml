```python
df['who_is_it_good_for'] = df['who_is_it_good_for'].apply(lambda x: [word.strip() for word in eval(x) if word.strip()])
expanded_df = df['who_is_it_good_for'].apply(pd.Series)
expanded_df.columns = [f'skin_problem_{i+1}' for i in range(expanded_df.shape[1])]
df = pd.concat([df.drop(columns=['who_is_it_good_for']), expanded_df], axis=1)
```

```python
df.head(2)
```

```python
# Assuming 'df' has the existing structure
# Reshape the DataFrame so that each skin_problem is stacked vertically
long_df = df.melt(id_vars=['name', 'short_description', 'what_is_it', 'what_does_it_do', 'url'], 
                  value_vars=[f'skin_problem_{i+1}' for i in range(10)], 
                  var_name='skin_problem_index', 
                  value_name='skin_problem')

# Drop the rows where 'skin_problem' is NaN
long_df = long_df.dropna(subset=['skin_problem'])

# Reset the index after melting
long_df = long_df.reset_index(drop=True)
```
